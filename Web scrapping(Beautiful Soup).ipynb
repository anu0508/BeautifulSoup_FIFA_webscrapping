It helps you understand how to perform web scrapping( which is extracting data from web page) and use that for further analysis. in this example our goal is to fetch all the team names participating in FIFA world cup 2018 from wikipedia page.

```python
#import requests as Requests allows you to send organic, grass-fed HTTP/1.1 requests, without the need for manual labor.
import requests
```

```python
# URL is saved in a variable
url='https://en.wikipedia.org/wiki/2018_FIFA_World_Cup_seeding'
```

```python
# in this below code i used requests.get method to save HTML page response in order to view the page we have use another methd
# called .text
website_url= requests.get(url)
```

```python
website_url.text
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(website_url.text,'html.parser')
print(soup.prettify())
```

```python
# after getting the page content I see that countries name are under class wikitable. so in this below code I used find method and passed class as 'wikitable'
# also since page has a table in which all the countries are listed, we have passed table.

My_table = soup.find('table',{'class':'wikitable'})
```

```python
# to get all the links containing football team name
links= My_table.findAll('a')
```

```python
links
```

```python
### in the links above we have all the teams played FIFA 2018. now we want to extract only team names from the links.
```

```python
countries=[]
for name in links:
    countries.append(name.get('title'))
    
print(countries)
```

```python
import pandas as pd
country=pd.DataFrame()
```

```python
country['Football_team_name']=countries
```
